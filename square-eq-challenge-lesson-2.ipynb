Lets plot two square equations one with real roots and another with imaginary roots:

```python
import numpy as np
import matplotlib.pyplot as plt

# Equation 1: x^2 - 5x + 6 = 0
eq1_coeffs = [1, -5, 5]

# Equation 2: x^2 + 4 = 0
eq2_coeffs = [1, 0, 4]

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values for equation 1
y1 = np.polyval(eq1_coeffs, x)

# Calculate y values for equation 2
y2 = np.polyval(eq2_coeffs, x)

# Plotting the parabolas
plt.plot(x, y1, label='x^2 - 5x + 6 = 0')
plt.plot(x, y2, label='x^2 + 4 = 0')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Set plot title and labels
plt.title('Parabolas')
plt.xlabel('x')
plt.ylabel('y')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
